### 7-3．深層学習のためのデータ準備
#### 全記事の分かち書き

```python
# Janomeのインストール
# !pip install janome
```

```python
import os
import re
from janome.tokenizer import Tokenizer

# 記事フォルダの指定
dirs = ['it-life-hack', 'movie-enter']

# 単語とラベルを格納するリストを生成
wakati = [] # 説明変数
label = [] # 目的変数

# 分かち書きオブジェクト生成
t = Tokenizer(wakati=True)

# 各フォルダのファイルを1つずつ読み込んで表示
for i, d in enumerate(dirs):
    # ファイルの取得
    files = os.listdir('./data/' + d)
    
    for file in files:
        # ファイルオープンと読み込み
        f = open('./data/' + d + '/' + file, 'r', encoding='utf-8')
        text = f.read()
        
        # 正規表現で不要な文字列を除去して表示
        reg_text = re.sub(r'[0-9a-zA-Z]+', '', text)
        reg_text = re.sub(r'[:;/+\.-]', '', reg_text)
        reg_text = re.sub(r'[\s\n]', '', reg_text)
        
        # 分かち書きして単語をリストへ格納
        wakati.append(t.tokenize(reg_text))
        
        # 記事ごとにラベルの格納
        label.append(i)
        
        # ファイルクローズ
        f.close()
```

```python
# リストのサイズ確認
print(len(wakati))
# 1つ目の記事の単語を確認
print(wakati[0])
# 1つ目の記事のラベルを確認
print(label[0])
```

#### 単語の数値化

```python
print(wakati)
```

```python
import itertools
from collections import Counter
import pandas as pd

# 単語の出現数をカウント
print(list(itertools.chain(* wakati)))
print(len(list(itertools.chain(* wakati))))
word_freq = Counter(itertools.chain(* wakati))

# 単語を出現数降順で並べ替え
dic = []
for word_uniq in word_freq.most_common():
    dic.append(word_uniq[0])

# 単語数を確認
# print(len(dic))
# 出現数の多い単語を5つ確認
# display(pd.DataFrame(dic))
```

```python
# 単語にIDを付与し辞書を作成
dic_inv = {}
for i, word_uniq in enumerate(dic, start=1):
    dic_inv.update({word_uniq: i})
    # if i == 10:
    #     break
    
# 辞書の長さを確認
print(len(dic_inv))
# print(dic_inv)
```

```python
# 練習問題4
# 単語をIDへ変換
wakati_id = [ [ dic_inv[word] for word in waka ] for waka in wakati ]

# リストのサイズ確認
print(len(wakati_id))
# 1つ目の記事の単語IDを確認
print(wakati_id[0])
```
