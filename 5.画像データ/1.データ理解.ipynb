```python
# OpenCVとMatplotlibの読み込み
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# 画像ファイルの読み込み
img = cv2.imread('./data/ants/swiss-army-ant.jpg')

# 画像のサイズの確認
print(img.shape)

# 画像ファイルの表示
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

#### ピクセル値の確認（カラー画像）

```python
# 画像のピクセル値を確認
print(img)
```

```python
# 画像の配列サイズを確認
print(len(img)) # 縦サイズ
print(len(img[0])) # 横サイズ
print(len(img[0][0])) # カラーチャンネル
```

```python
# Pandasの読み込み
import pandas as pd

# ピクセル配列をB、G、Rごとに分割
b, g, r = cv2.split(img)

# B配列をデータフレーム形式へ変換
b_df = pd.DataFrame(b)

# データサイズの確認
print(b_df.shape)
# データの先頭5行を表示
b_df.head()
```

```python
# 練習問題1
# G配列をデータフレーム形式へ変換
g_df = pd.DataFrame(g)

# データサイズの確認
print(g_df.shape)
# データの先頭5行を表示
g_df.head()
```

```python
# R配列をデータフレーム形式へ変換
r_df = pd.DataFrame(r)

# データサイズの確認
print(r_df.shape)
# データの先頭5行を表示
r_df.head()
```

#### ピクセル値の確認（グレースケール画像）

```python
# グレースケール画像へ変換
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 画像のサイズの確認
print(gray_img.shape)

# 画像ファイルの表示
plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
# 練習問題2
# 画像のピクセル値を確認
print(gray_img)
```

```python
# 練習問題3
# 画像の配列サイズを確認
print(len(gray_img)) # 縦サイズ
print(len(gray_img[0])) # 横サイズ
```

```python
# 配列をデータフレーム形式へ変換
gr_df = pd.DataFrame(gray_img)

# データサイズの確認
print(gr_df.shape)
# データの先頭5行を表示
gr_df.head()
```

#### ピクセル値の確認（2値化画像）

```python
# 2値化画像へ変換
ret, bin_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)

# 画像ファイルの表示
plt.imshow(bin_img, cmap='gray')
plt.show()
```

```python
# 画像のピクセル値を確認
print(bin_img)
```

```python
# 配列をデータフレーム形式へ変換
bin_df = pd.DataFrame(bin_img)

# データサイズの確認
print(bin_df.shape)
# データの先頭5行を表示
bin_df.head()
```
