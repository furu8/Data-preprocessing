### 3-1．データ理解
#### データの読み込み

```python
# Pandasの読み込み
import pandas as pd

# 任意のCSV
csv1 = "data/gps_log.csv"
csv2 = "data/bank.csv"
```

```python
# ファイルの読み込み
df = pd.read_csv(csv1, sep=',')
bank_df = pd.read_csv(csv2, sep=',')

# 先頭から5行目まで表示
df.head()
# bank_df.head()
```

```python
# 末尾から10行目まで表示
df.tail(10)
```

```python
# データの行数・列数を確認
print(df.shape)
```

```python
# データ型を確認
print(df.dtypes)
```

### 欠損値の確認

```python
# データ行に欠損値が含まれるか
print(bank_df.isnull().any(axis=1))

# データ列に欠損値が含まれるか
print(bank_df.isnull().any(axis=0))
```

```python
# データ行ごとの欠損値の個数
print(df.isnull().sum(axis=1))

# データ列ごとの欠損値の個数
print(df.isnull().sum(axis=0))
```

```python
# データ行ごとの欠損値の個数
print(df.isnull().sum(axis=1).sort_values(ascending=False))
```

### 統計量の計算

```python
# データ型が数値項目の各種統計量の計算
df.describe()
```

```python
# データ型が文字列項目の各種統計量の計算
df.describe(include=[object])
```

### データの可視化１

```python
# Matplotlibの読み込み
import matplotlib.pyplot as plt
%matplotlib inline

# ageのヒストグラムを作成
plt.hist(df['altitude'])
# x軸（横軸）とy軸（縦軸）のラベルを追加
plt.xlabel('altitude')
plt.ylabel('freq')
plt.show()
```

```python
# 散布図作成
plt.scatter(df['latitude'], df['longitude'])
# x軸（横軸）とy軸（縦軸）のラベルを追加
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.show()

# 相関係数
print(df[['latitude', 'longitude']].corr())
```

```python
# 散布図行列の描画
pd.plotting.scatter_matrix(df[['altitude','horizontal_accuracy','speed','course']])
# plt.tight_layout()
plt.show()
```

```python
# objectの比率を計算（値しかないデータには不適用）
print(bank_df['job'].value_counts(ascending=False, normalize=True))
```

```python
# ラベル
label = bank_df['job'].value_counts(ascending=False, normalize=True).index
# 値
vals = bank_df['job'].value_counts(ascending=False, normalize=True).values

# 円グラフを作成
plt.pie(vals, labels=label)
plt.axis('equal')
plt.show()
```

### データの可視化２

```python
# yがyesとnoのときのageデータを抽出
y_yes = bank_df[bank_df['y'] == 'yes']
y_no = bank_df[bank_df['y'] == 'no']
# yesとnoのデータをまとめる
y_age = [y_yes['age'], y_no['age']]

# 箱ひげ図の描画
plt.boxplot(y_age)
# x軸（横軸）とy軸（縦軸）のラベルを追加
plt.xlabel('y')
plt.ylabel('age')
ax = plt.gca()
# y軸（縦軸）のラベルを追加
plt.setp(ax, xticklabels = ['yes','no'])
plt.show()
```

```python
# 練習問題7
# yesとnoのデータをまとめる
y_balance = [y_yes['balance'], y_no['balance']]

# 箱ひげ図の描画
plt.boxplot(y_balance)
# x軸（横軸）とy軸（縦軸）のラベルを追加
plt.xlabel('y')
plt.ylabel('balance')
ax = plt.gca()
# y軸（縦軸）のラベルを追加
plt.setp(ax, xticklabels = ['yes','no'])
plt.show()
```
