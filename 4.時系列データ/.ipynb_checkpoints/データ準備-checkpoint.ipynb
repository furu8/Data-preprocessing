### 6-1．データ理解
#### データの読み込み

```python
import pandas as pd

# ファイルの読み込み
gps_df = pd.read_csv('data/gps_log.csv', sep=',')


# 先頭から5行目まで表示
gps_df.head()
```

```python
# データの行数・列数を確認
print(gps_df.shape)
```

```python
# データ型を確認
print(gps_df.dtypes)
```

```python
# dateをobject型からdatetime型へ変換
gps_df['time'] = pd.to_datetime(gps_df['time'], format='%Y-%m-%d %H:%M:%S')

# date列のデータ型を確認
print(gps_df['time'].dtypes)
# date列1行目のデータ型を確認
print(type(gps_df['time'][0]))
```

```python
# 前行との時間差を計算
gps_df['gps_sec'] = gps_df['time'].diff().dt.total_seconds()
# 先頭要素の欠損を0埋め
gps_df['gps_sec'] = gps_df['gps_sec'].fillna(0)
# 先頭から5行目まで表示
gps_df['gps_sec'].head()
```

```python
# 経過時間（秒）の計算
gps_df['cum_sec'] = gps_df['gps_sec'].cumsum()

# 元の日時と経過時間（秒）を表示
gps_df[['time', 'cum_sec']].head()
```

```python
# 経過時間（分）の計算
gps_df['cum_min'] = (gps_df['cum_sec']/60).round(2).astype(int)

# 1時間単位の連番を確認
print(gps_df['cum_min'].unique())

# 元の日時と経過時間（分）を表示
gps_df[['time', 'cum_sec', 'cum_min']].head(30)
```

#### 欠損値の確認

```python
# データ行ごとの欠損値の個数
print(gps_df.isnull().sum(axis=1).sort_values(ascending=False))

# データ列ごとの欠損値の個数
print(gps_df.isnull().sum(axis=0))
```

#### 統計量の計算

```python
# データ型が数値項目の各種統計量の計算
gps_df.describe()
# gps_df = gps_df.drop("gps_min", axis=1)
# gps_df.describe()
```

```python
# 各項目の平均値
gps_df.groupby('cum_min').mean()
```

```python
# 各項目の標準偏差
gps_df.groupby('cum_min').std()
```

#### データの可視化

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Appliancesの折れ線グラフ
plt.plot(gps_df['cum_sec'], gps_df['altitude'])

# y軸（縦軸）のラベルを追加
plt.xlabel('cum_sec')
plt.ylabel('altitude')
plt.show()
```

```python
# T1～T3の折れ線グラフ
plt.plot(gps_df['cum_sec'], gps_df['latitude'], '-r', label='latitude')
plt.plot(gps_df['cum_sec'], gps_df['longitude'], '-g', label='longitude')
plt.plot(gps_df['cum_sec'], gps_df['altitude'], '-b', label='altitude')

# ラベルを追加
plt.xlabel('cum_sec')
plt.legend()
plt.show()
```
