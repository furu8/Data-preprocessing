### 7-1．データ理解
#### Janomeを使った形態素解析

```python
# ファイルオープン
f = open('./data/it-life-hack/it-life-hack-6292880.txt', encoding='utf-8')

# ファイルの読み込み
text = f.read()
# テキストの確認
print(text)

# ファイルクローズ
f.close()
```

```python
# Janomeのインストール
# !pip install janome
```

```python
from janome.tokenizer import Tokenizer

# オブジェクト生成
t = Tokenizer()
# 形態素解析
for token in t.tokenize(text):
    print(token)
```

#### テキストの正規化

```python
import re

# 半角数字0～9、小文字大文字a～ZなどURLに含まれる文字列を除去
reg_text = re.sub(r'[0-9a-zA-Z]+', '', text)
reg_text = re.sub(r'[:;/+\.-]', '', reg_text)

print(reg_text)
```

```python
# 練習問題1
reg_text = re.sub(r'[\s\n]', '', reg_text)

print(reg_text)
```

#### 単語を品詞で抽出

```python
from janome.analyzer import Analyzer
from janome.tokenfilter import POSKeepFilter

# 名詞の単語のみ抽出するフィルターを生成
token_filters = [POSKeepFilter(['名詞'])]
# フィルターを搭載した解析フレームワークの生成
a = Analyzer([], t, token_filters)

# フィルタ付き形態素解析
for token in a.analyze(reg_text):
    print(token)
```

#### 単語の出現回数をカウント

```python
# 練習問題2
import collections

# 単語格納リスト
text = []

# フィルタ付き形態素解析
for token in a.analyze(reg_text):
    text.append(token.surface)

# 単語の出現回数をカウント
c = collections.Counter(text)
print(c)
```
